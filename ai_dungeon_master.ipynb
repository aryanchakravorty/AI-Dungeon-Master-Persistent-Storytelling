```python
# Standard library and Google imports
from google.colab import userdata   # Access Colab-stored credentials or API keys
import os                           # For system-level operations like file paths
import uuid                         # For generating unique identifiers
import google.generativeai as genai  # Google's Generative AI SDK
from typing import Annotated
from typing_extensions import TypedDict
from IPython.display import Image, display, Markdown  # For displaying rich outputs in Colab
import pandas as pd                 # Data handling and manipulation
from datetime import datetime       # For timestamps and logging

# --- LangChain/LangGraph Imports ---

# Integrations with embeddings, models, and external tools
from langchain_chroma import Chroma                            # Vector store for embeddings
from langchain_community.embeddings import HuggingFaceEmbeddings  # Generate text embeddings
from langchain_google_genai import ChatGoogleGenerativeAI          # Connects LangChain with Google's Gemini models
from langchain_tavily import TavilySearch                         # Enables web or knowledge base search

# Core LangChain/LangGraph components for building conversational graphs
from langchain_core.runnables.config import RunnableConfig
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage, RemoveMessage  # Message types in conversation flow
from langgraph.graph import StateGraph, START, END                # Defines structure and flow of the AI reasoning graph
from langgraph.graph.message import add_messages, MessagesState   # For managing message states
from langgraph.prebuilt import ToolNode, tools_condition          # To integrate and control external tools
from langgraph.checkpoint.memory import MemorySaver               # For saving conversation state and checkpoints
```

```python
# Initializing the Google Generative AI chat model (Gemini 2.0 Flash) for conversational processing
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
# Creating a Tavily search tool instance, limiting it to 3 search results per query
tool = TavilySearch(max_results=3)

# Storing the tool inside a list for easier management and integration with the graph later
tools = [tool]
```

```python
# Setting up the embedding model using Hugging Face's 'Snowflake Arctic' for generating text embeddings
embedding_model = HuggingFaceEmbeddings(model_name="Snowflake/snowflake-arctic-embed-m")

# Creating a Chroma vector database to store and retrieve embeddings efficiently
vector_db = Chroma(
    collection_name="memory",                   # Name of the collection for organized storage
    embedding_function=embedding_model,         # Embedding model used to convert text into vectors
    persist_directory="./chroma_langchain_db"   # Directory to persist the database locally
)
```

```python
# Binding the Tavily search tool to the language model, enabling the model to call the tool when needed
model = llm.bind_tools(tools)
```

```python
# Defining a function to extract and store key facts from user messages into the Chroma database
def store_facts_to_chroma(state: MessagesState, config: RunnableConfig):
    """
    Capturing important user facts, summarizing them, and storing them for memory retention.
    """

    user_id = config["configurable"]["user_id"]       # Getting the user ID from configuration
    last_msg = state["messages"][-1]                  # Fetching the most recent message from the conversation

    # Skipping processing if the last message is from the AI, since we only want user-provided facts
    if isinstance(last_msg, AIMessage):
        return {"messages": state["messages"]}

    # Creating a prompt asking the model to summarize the user's message into a clear factual statement
    summarization_prompt = HumanMessage(
        content=(
            "Read the user's latest message and extract a single concise factual statement. "
            "Do NOT include any explanations, opinions, or conversational text. "
            "Output only the factual statement, in one clear sentence."
        )
    )

    try:
        # Using the model to generate a summarized version of the user's statement
        summary_response = model.invoke([last_msg, summarization_prompt])
        summarized_fact = summary_response.content.strip()

        if summarized_fact:
            # Determining the type of fact (default category is 'event')
            fact_type = determine_fact_type(summarized_fact)

            # Converting the summarized text into embeddings for vector storage
            embeddings = embedding_model.embed_documents([summarized_fact])

            # Storing the summarized fact into the Chroma vector database with metadata
            vector_db.add_texts(
                texts=[summarized_fact],
                metadatas=[{
                    "user": user_id,
                    "type": fact_type,
                    "timestamp": str(pd.Timestamp.now())
                }],
                ids=[str(uuid.uuid4())]   # Assigning a unique ID to each stored fact
            )
    except Exception as e:
        # Logging any errors that occur during summarization or storage
        print(f"Error summarizing fact: {e}")

    # Returning the message state back unchanged after processing
    return {"messages": state["messages"]}
```

```python
# ---------- Adaptive RAG (Retrieval-Augmented Generation) Routing ----------

from typing import List

# Defining a router function to decide the processing path for a user query
def route_query(query: str) -> str:
    """
    Determining whether a user's question needs memory retrieval (vectorstore)
    or can be answered directly via chat.
    """

    # System instruction guiding the model to make a routing decision
    system = (
        "You are a router. Determine if the user's question requires factual retrieval from memory "
        "(vectorstore) or if it is a creative/narrative request that does NOT require retrieval (chat). "
        "Respond ONLY with 'vectorstore' or 'chat'. Do not add anything else."
    )


    # Invoking the model to classify the query
    resp = model.invoke([SystemMessage(content=system), HumanMessage(content=query)])
    result = resp.content.strip().lower()

    # Returning the routing decision: default to 'chat' if 'vectorstore' isn't mentioned
    return "vectorstore" if "vectorstore" in result else "chat"
```

```python
# Defining a function to rephrase user queries for better vector search performance
def rewrite_for_retrieval(query: str) -> str:
    """
    Rewriting the user's question to improve results when searching the vector database.
    """

    # System instruction telling the model to focus on keywords and preserve the intent
    sys = (
        "Rewrite the user's question for vector database retrieval. Be concise, preserve the original intent, "
        "and emphasize the most important keywords. Output ONLY the rewritten query with no extra text."
    )


    # Using the model to generate an optimized version of the query
    resp = model.invoke([SystemMessage(content=sys), HumanMessage(content=query)])

    # Returning the rewritten query
    return resp.content.strip()
```

```python
# Defining a function to prepare retrieved documents for use in prompts
def format_docs_for_prompt(docs: List) -> str:
    """
    Formatting a list of retrieved documents into a single string
    that can be inserted into a prompt for the model.
    """

    # Joining the content of each document with double line breaks for readability
    return "\n\n".join(getattr(d, "page_content", str(d)) for d in docs)
```

```python
# Defining a function to pull the NPC's name from structured text
def extract_npc_name(npc_text: str) -> str:
    """
    Extracting the NPC's name from a block of text formatted with 'NPC Name:'.
    """

    # Splitting the text into individual lines
    lines = npc_text.split("\n")

    # Searching for the line that contains the NPC's name
    for line in lines:
        if line.startswith("NPC Name:"):
            # Removing the label and any extra spaces, returning just the name
            name = line.replace("NPC Name:", "").strip()
            return name.split("\n")[0].strip()

    # Returning None if no NPC name is found
    return None
```

```python
# Defining a function to classify a fact into a specific type for better organization
def determine_fact_type(text: str) -> str:
    """
    Categorizing a piece of information into one of several types:
    'quest', 'event', 'relationship', 'fact', or 'discovery'.
    """

    text_lower = text.lower()  # Converting text to lowercase for easier matching

    # Checking for keywords to assign the appropriate fact type
    if any(word in text_lower for word in ["quest", "mission", "task", "objective"]):
        return "quest"
    elif any(word in text_lower for word in ["like", "hate", "love", "trust", "betray"]):
        return "relationship"
    elif any(word in text_lower for word in ["find", "discover", "found", "learned", "secret"]):
        return "discovery"
    elif any(word in text_lower for word in ["happen", "did", "was", "event", "occurred"]):
        return "event"
    else:
        # Default type if no keywords match
        return "fact"
```

```python
# Defining a function to find mentions of known NPCs in user input
def extract_npc_mentions(user_input: str, known_npcs: List[str]) -> List[str]:
    """
    Identifying which NPCs from a predefined list are mentioned
    in the user's input to track references accurately.
    """

    mentioned = []
    user_input_lower = user_input.lower()  # Making the comparison case-insensitive

    # Checking each known NPC to see if their name appears in the input
    for npc in known_npcs:
        if npc.lower() in user_input_lower:
            mentioned.append(npc)

    # Returning the list of NPCs that were mentioned
    return mentioned
```

```python
# Defining a function to fetch all NPC names previously stored for a specific user
def get_known_npcs(user_id: str) -> List[str]:
    """
    Retrieving all NPC names associated with the given user from the vector database.
    """

    try:
        # Querying the vector database for entries of type 'npc' for this user
        results = vector_db.get(
            where={"user": user_id, "type": "npc"}
        )

        npc_names = []

        # Extracting NPC names from the returned metadata
        if results and results.get("metadatas"):
            for metadata in results["metadatas"]:
                if "npc_name" in metadata:
                    npc_names.append(metadata["npc_name"])

        # Removing duplicate names and returning the list
        return list(set(npc_names))

    except Exception:
        # Returning an empty list if any error occurs during retrieval
        return []
```

```python
# Defining a function to log player actions and NPC responses for memory tracking
def store_npc_action(
    user_input: str,
    dm_response: str,
    npc_name: str,
    user_id: str
):
    """
    Recording interactions between the player and an NPC.
    Stores what the player did and how the NPC responded for future reference.
    """

    try:
        # Creating a structured record of the interaction
        action_record = (
            f"NPC: {npc_name}\n"
            f"Player Action: {user_input}\n"
            f"NPC Response/Outcome: {dm_response}\n"
        )

        # Generating a unique ID for this interaction
        action_id = str(uuid.uuid4())

        # Saving the record in the vector database with relevant metadata
        vector_db.add_texts(
            texts=[action_record],
            metadatas=[{
                "user": user_id,
                "type": "npc_action",
                "npc_name": npc_name,
                "timestamp": str(pd.Timestamp.now())
            }],
            ids=[action_id]
        )

    except Exception as e:
        # Logging any error that occurs during storage
        print(f"Error storing NPC action: {e}")
```

```python
# Defining a function to identify the main NPC mentioned in a DM's response
def extract_npc_name_from_dm_response(dm_response: str) -> str:
    """
    Parsing the DM's response to find which NPC was primarily involved.
    Looks for bolded text patterns (markdown **NPC Name**) as indicators.
    """

    import re

    # Searching for text enclosed in double asterisks (markdown bold)
    bold_matches = re.findall(r'\*\*([^*]+)\*\*', dm_response)

    # Returning the first match, assuming it's the NPC's name
    if bold_matches:
        return bold_matches[0]

    # Return None if no bolded text is found
    return None
```

```python
# Defining a function to gather full context for one or more NPCs for DM use
def retrieve_npc_context(npc_names: List[str], user_id: str) -> str:
    """
    Collecting NPC profiles and recent action history.
    Returns a formatted string that the DM can use to generate context-aware responses.
    """

    if not npc_names:
        return ""

    npc_context_parts = []

    for npc_name in npc_names:
        try:
            # Fetching the NPC's main profile from the vector database
            npc_profile_results = vector_db.similarity_search_with_score(
                f"NPC {npc_name} profile",
                k=1,
                filter={"user": user_id, "type": "npc", "npc_name": npc_name}
            )

            npc_profile = ""
            if npc_profile_results:
                doc, score = npc_profile_results[0]
                npc_profile = doc.page_content

            # Fetching recent player interactions with this NPC
            action_history_results = vector_db.similarity_search(
                f"{npc_name} action interaction",
                k=5,
                filter={"user": user_id, "type": "npc_action", "npc_name": npc_name}
            )

            action_history = ""
            if action_history_results:
                actions = [doc.page_content for doc in action_history_results]
                action_history = "\n---\n".join(actions)

            # Combining profile and action history into a single section
            npc_section = f"## {npc_name}\n\n{npc_profile}"
            if action_history:
                npc_section += f"\n\n### Interaction History:\n{action_history}"

            npc_context_parts.append(npc_section)

        except Exception as e:
            # Logging any issues retrieving data for a specific NPC
            print(f"Error retrieving context for {npc_name}: {e}")
            continue

    # Returning all NPC contexts combined as a single string
    return "\n\n".join(npc_context_parts)
```

```python
# Defining a function to fetch short summaries of all NPCs for a user
def retrieve_all_npc_summaries(user_id: str) -> str:
    """
    Gathering brief NPC summaries to provide overall game context.
    Helps the DM maintain consistency in the world and characters.
    """

    try:
        # Searching for up to 20 NPC profiles for this user
        results = vector_db.similarity_search(
            "NPC profile character",
            k=20,
            filter={"user": user_id, "type": "npc"}
        )

        # Return a message if no NPCs are found
        if not results:
            return "No NPCs registered yet."

        # Extracting the text content from each document and joining with separators
        summaries = [doc.page_content for doc in results]
        return "\n\n---\n\n".join(summaries)

    except Exception:
        # Return fallback message if retrieval fails
        return "No NPC data available."
```

```python
# Defining a function to count player interactions with a specific NPC
def get_npc_interaction_count(npc_name: str, user_id: str) -> int:
    """
    Tracking how many times the player has interacted with a given NPC.
    Helps gauge relationship depth or engagement level with the character.
    """

    try:
        # Retrieving all action records for this NPC and user
        results = vector_db.get(
            where={
                "user": user_id,
                "type": "npc_action",
                "npc_name": npc_name
            }
        )

        # Returning the number of interactions found
        return len(results.get("ids", [])) if results else 0

    except Exception:
        # Return 0 if any error occurs
        return 0
```

```python
# Defining a function to evaluate if a retrieved document is relevant to a user's question
def grade_retrieval(question: str, doc_text: str) -> bool:
    """
    Asking the model to determine whether a given document helps answer the question.
    Returns True if relevant, False otherwise.
    """

    # System instruction telling the model to respond with a simple yes or no
    sys = (
        "Is this document relevant to the question? Answer 'yes' or 'no' only."
    )

    # Invoking the model to evaluate the relevance of the document
    resp = model.invoke([
        SystemMessage(content=sys),
        HumanMessage(content=f"Question: {question}\n\nDocument:\n{doc_text}")
    ])

    # Returning True if the model answered yes, False otherwise
    return resp.content.strip().lower().startswith("y")
```

```python
# Defining a function to check if the AI's generation is based on retrieved facts
def grade_hallucination(docs_text: str, generation_text: str) -> bool:
    """
    Asking the model whether its response is consistent with the provided factual documents.
    Returns True if the generation is grounded, False if it contains hallucinations.
    """

    # System instruction to ensure the model focuses on factual consistency
    sys = (
        "Check if this generation is fully grounded in and consistent with the provided facts. "
        "Respond with 'yes' or 'no' ONLY, without any explanation or extra text."
    )

    # Invoking the model to assess whether the generated text matches the facts
    resp = model.invoke([
        SystemMessage(content=sys),
        HumanMessage(content=f"Facts:\n{docs_text}\n\nGeneration:\n{generation_text}")
    ])

    # Returning True if the model answers 'yes', indicating the generation is factual
    return resp.content.strip().lower().startswith("y")
```

```python
# Defining a function to verify if the AI's answer directly addresses the user's question
def grade_answer(question: str, answer_text: str) -> bool:
    """
    Evaluates whether the provided answer actually responds to the question.
    Returns True if it addresses the question, False otherwise.
    """

    # System instruction to focus on answer relevance
    sys = (
        "Check if this answer directly addresses the question. "
        "Respond with 'yes' or 'no' ONLY, without any explanations or extra text."
    )

    # Invoking the model to check if the answer is on-topic and relevant
    resp = model.invoke([
        SystemMessage(content=sys),
        HumanMessage(content=f"Question: {question}\n\nAnswer:\n{answer_text}")
    ])

    # Returning True if the model indicates 'yes', otherwise False
    return resp.content.strip().lower().startswith("y")
```

```python
# Defining a function to initialize the game world, generating story and NPCs
def story_init_node(state: MessagesState, config: RunnableConfig):
    """
    Setting up the fantasy game world on demand.
    Generates the initial storyline, creates NPCs, and stores all data in memory.
    Returns initialization details to be used by the chat node.
    """

    user_id = config["configurable"]["user_id"]  # Getting the current user's ID

    # Prompting the model to generate the story setting and 4 NPCs with details
    init_prompt = HumanMessage(
        content=(
            "You are a fantasy world builder. Create an engaging dungeon adventure scenario.\n"
            "Generate the following:\n"
            "1. A concise (2-3 sentence) setting description that sets mood and stakes.\n"
            "2. Exactly 4 NPCs in this format:\n"
            "   NPC Name: [name]\n"
            "   Role: [role]\n"
            "   Personality: [brief personality traits]\n"
            "   Backstory: [2-3 sentence background story]\n"
            "   Current Location: [where they are]\n"
            "   Relationship to Player: [initial relationship or attitude]\n"
            "\n"
            "Ensure the setting and NPCs are immersive, distinctive, and memorable."
        )
    )

    try:
        # Invoking the model to generate the initial story content
        init_response = model.invoke([init_prompt])
        storyline_content = init_response.content

        # Storing the main storyline in the vector database
        storyline_id = str(uuid.uuid4())
        vector_db.add_texts(
            texts=[f"STORYLINE_START:\n{storyline_content}"],
            metadatas=[{
                "user": user_id,
                "type": "storyline",
                "timestamp": str(pd.Timestamp.now())
            }],
            ids=[storyline_id]
        )

        # Extracting and storing each NPC separately
        npc_sections = storyline_content.split("NPC Name:")
        for section in npc_sections[1:]:
            npc_text = "NPC Name:" + section
            npc_name = extract_npc_name(npc_text)

            if npc_name:
                npc_id = str(uuid.uuid4())
                vector_db.add_texts(
                    texts=[npc_text],
                    metadatas=[{
                        "user": user_id,
                        "type": "npc",
                        "npc_name": npc_name,
                        "timestamp": str(pd.Timestamp.now())
                    }],
                    ids=[npc_id]
                )

        # Returning initialization status and content
        return {
            "messages": state["messages"],
            "story_initialized": True,
            "storyline_content": storyline_content
        }

    except Exception as e:
        # Handling errors during story creation
        print(f"Error in story initialization: {e}")
        return {
            "messages": state["messages"],
            "story_initialized": False,
            "storyline_content": None
        }
```

```python
# Defining the main chat node that drives the game flow and interacts with the player
def chat_node(state: MessagesState, config: RunnableConfig):
    """
    Enhanced chat node that manages the fantasy adventure.
    Initializes the story when needed, retrieves context, adapts to player queries,
    and stores NPC interactions for consistent world-building.
    """

    user_id = config["configurable"]["user_id"]
    messages = state["messages"]
    last_input = messages[-1]
    history = messages[:-1]
    short_history = history[-10:]  # Using last 10 messages for context

    question_text = last_input.content

    # ============ STEP 0: CHECK IF STORY INITIALIZED ============
    story_initialized = state.get("story_initialized", False)
    storyline_content = state.get("storyline_content", None)

    # Initialize the world if not already done
    if not story_initialized:
        try:
            existing_memories = vector_db.similarity_search(
                "storyline character npc",
                k=1,
                filter={"user": user_id, "type": "storyline"}
            )
            if not existing_memories:
                init_result = story_init_node(state, config)
                story_initialized = init_result.get("story_initialized", False)
                storyline_content = init_result.get("storyline_content", None)

                # Present initial world to player
                if story_initialized and storyline_content:
                    welcome_msg = AIMessage(
                        content=f"🎭 **Welcome to Your Adventure**\n\n{storyline_content}\n\n---\n\nWhat do you do?"
                    )
                    return {
                        "messages": messages + [welcome_msg],
                        "story_initialized": True,
                        "storyline_content": storyline_content
                    }
            else:
                story_initialized = True
        except Exception:
            pass

    # ============ STEP 1: GET KNOWN NPCs ============
    known_npcs = get_known_npcs(user_id)
    mentioned_npcs = extract_npc_mentions(question_text, known_npcs)

    # ============ STEP 2: RETRIEVE DOCUMENTS + NPC CONTEXT ============
    rewritten_query = question_text
    try:
        rewritten_query = rewrite_for_retrieval(question_text)
    except Exception:
        pass

    docs_text = ""
    npc_context = ""

    try:
        # Standard memory retrieval
        retrieved_docs = vector_db.similarity_search_with_score(
            rewritten_query,
            k=5,
            filter={"user": user_id}
        )
        docs_only = [d for d, _ in retrieved_docs]
        docs_text = format_docs_for_prompt(docs_only) if docs_only else ""
    except Exception:
        docs_only = []
        docs_text = ""

    # NPC-specific retrieval
    if mentioned_npcs:
        try:
            npc_context = retrieve_npc_context(mentioned_npcs, user_id)
        except Exception:
            npc_context = ""

    # ============ STEP 3: BUILD ENHANCED SYSTEM PROMPT ============
    long_term_context = docs_text if docs_text else "No character facts yet."

    npc_section = ""
    if npc_context:
        npc_section = f"\n\n🎭 **NPC Context:**\n{npc_context}"
    elif known_npcs:
        all_npcs = retrieve_all_npc_summaries(user_id)
        npc_section = f"\n\n🎭 **Known NPCs:**\n{all_npcs}"

    guidance_prompt = (
        "You are a **Dungeon Master (DM)** in a text-based fantasy role-playing game.\n"
        "Your goals:\n"
        "- Write in **Markdown** for readability.\n"
        "- Narrate in a cinematic, engaging style.\n"
        "- Use **bold** for key names, items, or dramatic moments.\n"
        "- Break narration into **short paragraphs or bullet points** — avoid long single blocks of text.\n"
        "- Use **line breaks** to separate narration from choices.\n"
        "- Recall facts and events from long-term memory (summarized below).\n"
        "- Use NPC context to maintain consistent character behavior and relationships.\n"
        "- Prioritize the player's actions, choices, and requirements.\n"
        "- Present clear choices and respond creatively to player decisions.\n"
        "- Keep tone immersive, descriptive, and cinematic.\n"
        "- Always provide multiple choices to the player, but also allow free-form responses if the player types something outside the given options."
        "- If the player is silent or asks for a recap, provide a concise summary of the adventure so far.\n"
        "\n"
        f"📖 **Long-Term Memory:**\n{long_term_context}\n"
        f"{npc_section}"
    )


    system_message = SystemMessage(content=guidance_prompt)

    # ============ STEP 4: ROUTE QUERY ============
    route = "vectorstore"
    try:
        route = route_query(question_text)
    except Exception:
        route = "vectorstore"

    # ============ STEP 5: ADAPTIVE RAG PIPELINE ============
    final_generation = None

    if route == "vectorstore" and (docs_only or mentioned_npcs):
        try:
            combined_context = docs_text
            if npc_context:
                combined_context += f"\n\n{npc_context}"

            # Grade retrieval relevance
            retrieval_grade_ok = grade_retrieval(
                rewritten_query,
                combined_context[:4000]
            )

            if retrieval_grade_ok:
                # Generate response with RAG
                rag_system = SystemMessage(
                    content=(
                        system_message.content +
                        "\n\nRelevant facts from memory:\n" +
                        combined_context
                    )
                )
                messages_for_rag = [rag_system] + short_history + [last_input]

                try:
                    rag_response = model.invoke(messages_for_rag)
                    rag_generation = rag_response.content

                    # Check for hallucinations
                    hallucination_ok = grade_hallucination(
                        combined_context[:4000],
                        rag_generation
                    )

                    # Check if answer addresses question
                    answer_ok = grade_answer(question_text, rag_generation)

                    if hallucination_ok and answer_ok:
                        final_generation = rag_generation

                except Exception:
                    pass

        except Exception:
            pass

    # ============ STEP 6: CHAT_ONLY FALLBACK ============
    if final_generation is None:
        try:
            messages_for_chat = [system_message] + short_history + [last_input]
            chat_response = model.invoke(messages_for_chat)
            final_generation = chat_response.content
        except Exception:
            final_generation = "The DM is momentarily lost in thought... [Error]"

    # ============ STEP 7: STORE NPC INTERACTIONS ============
    if mentioned_npcs and final_generation:
        for npc_name in mentioned_npcs:
            try:
                store_npc_action(
                    user_input=question_text,
                    dm_response=final_generation,
                    npc_name=npc_name,
                    user_id=user_id
                )
            except Exception as e:
                print(f"Error storing NPC action for {npc_name}: {e}")

    # ============ STEP 8: RETURN UPDATED STATE ============
    ai_msg = AIMessage(content=final_generation)

    updated_messages = []
    removed_message_ids = {m.id for m in messages if isinstance(m, RemoveMessage)}
    for msg in messages + [ai_msg]:
        if not isinstance(msg, RemoveMessage) and msg.id not in removed_message_ids:
            updated_messages.append(msg)

    return {
        "messages": updated_messages,
        "story_initialized": story_initialized,
        "storyline_content": storyline_content
    }
```

```python
# ============ BUILD THE GRAPH ============

# Extending the message state to include story initialization info
class GameState(MessagesState):
    """
    Game state that tracks whether the story has been initialized
    and stores the initial storyline content.
    """
    story_initialized: bool = False
    storyline_content: str = None
```

```python
# ============ BUILD THE GRAPH ============

# 1. Create the graph builder using the custom GameState
builder = StateGraph(state_schema=GameState)  # Using extended state to track story

# 2. Add nodes to the graph
# store_facts node handles saving user facts to Chroma
builder.add_node("store_facts", store_facts_to_chroma)
# chat node manages the main game flow, including story initialization
builder.add_node("chat", chat_node)

# 3. Add a ToolNode to handle external tool calls (e.g., Tavily search)
tool_node = ToolNode(tools=[tool])
builder.add_node("tools", tool_node)

# ============ EDGE CONFIGURATION ============

# 4. Conditional edges from chat based on tool usage
builder.add_conditional_edges(
    "chat",
    tools_condition,
    {
        "tools": "tools",  # If tools are needed, route to tool node
        END: END,           # Otherwise, end the flow
    }
)

# 5. Edge from tools back to chat for continued processing
builder.add_edge("tools", "chat")

# 6. Simplified main flow: START → store_facts → chat → END
builder.add_edge(START, "store_facts")
builder.add_edge("store_facts", "chat")
builder.add_edge("chat", END)

# 7. Compile the graph with memory saver for checkpoints
memory = MemorySaver()
chat_app = builder.compile(checkpointer=memory)

print("✅ Graph compiled successfully!")
print("Graph structure: START → store_facts → chat (invokes story_init on first run)")
```
